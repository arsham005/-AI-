```python
from matplotlib import pyplot as plt
import cv2
```

```python
img = cv2.imread("download.webp")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
model = cv2.CascadeClassifier("model.xml")
```

```python
face = model.detectMultiScale(img)
```

```python
img.shape
```

```python
x = face[0][0]
y = face[0][1]
a = face[0][2]
b = face[0][3]
```

```python
img = cv2.rectangle(img , (x,y),(x + a,y + b),(250,0,0),3)
```

```python
plt.imshow(img)
```
